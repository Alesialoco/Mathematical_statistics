```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

1.jpg

```python
#Значения
n = 3
x = np.array([-1.11, -6.10, 2.42])
sigma_x_2 = 2

m = 2
y = np.array([-2.29, -2.91])
sigma_y_2 = 1

x_mean = x.mean()
y_mean = y.mean()

delta = (x_mean - y_mean) / (sigma_x_2 / n + sigma_y_2 / m) ** 0.5

# Уровень значимости
alpha = 0.05

print(f"Δ : {delta}")
```

$G : Δ \ge U_{1-\alpha/2}$

```python
p_value = 1 - norm.cdf(delta)
print(f"P-value: {p_value}")
```

$p-value (длина) \ge \alpha$

Нет оснований отвергнуть $H_0$

###График мощности

$W = 1 - P(x < U_{1-\alpha} - \frac{a-b}{\frac{\sigma_x}{n} + \frac{\sigma_y}{m}})$

```python
a_b = np.linspace(-1, 5, 100)
power = 1 - norm.cdf(norm.ppf(1 - alpha) - a_b / (sigma_x_2 / n + sigma_y_2 / m))

plt.figure(figsize=(9, 5))
plt.plot(a_b, power, color='purple')
plt.xlabel('a-b (разность средних)')
plt.ylabel('W')
plt.title('График мощности критерия')
plt.grid(linewidth=0.4)
plt.show()
```
